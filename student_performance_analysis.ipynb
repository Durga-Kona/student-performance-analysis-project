```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("student_academic_performance.csv")
data.head()
```

```python
sns.scatterplot(x="Study Time per Week", y="Math Score", data=data)
plt.title("Study Time per Week vs Final Score")
plt.show()
```

```python
corr = data.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```
